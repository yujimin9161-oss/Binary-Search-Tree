2) 현재 높이 4를 유지하면서 키를 추가한다면 가능한 리프 노드의 최대 개수는 몇 개인가?
답: 8개 / 높이가 $h$인 이진 트리에서 레벨 $L$에 가질 수 있는 최대 노드 수는 $2^{L-1}$이다. 높이 4를 유지하는 포화 이진 트리에서 리프 노드의 최대 개수는 $2^{4-1} = 2^3 = 8$개이다.

3) 높이 4인 이진 검색 트리가 포화 이진 트리 모양이 되면 리프 노드와 내부 노드의 수 차이는?
답: 1개 / 전체 노드 수: $2^4 - 1 = 15$개   
리프 노드 수: $2^{4-1} = 8$개    
내부 노드 수: $15 - 8 = 7$개     
따라서 $8 - 7 = 1$로 1개 차이가 난다.

4) 위 3)의 성질은 높이가 $h$인 모든 포화 이진 트리에 대해 성립하는가? 답: 성립함 / 리프 노드의 수 $n_L = 2^{h-1}$이고, 내부 노드의 수 $n_I = 2^{h-1} - 1$이므로, $n_L - n_I = 1$의 관계가 항상 유지된다.

1) 트리의 총 노드 수가 $n$이면 리프 노드의 수는 몇 개인가? 답: $\lceil n / 2 \rceil$개 / 이진 트리에서 자식 노드가 2개인 노드의 수를 $n_2$, 자식이 1개인 노드를 $n_1$, 리프 노드를 $n_0$라고 할 때, 전체 노드 수는 $n = n_2 + n_1 + n_0$이다.이진 트리의 성질에 의해 $n_0 = n_2 + 1$이 성립한다.완전 이진 트리에서는 자식이 1개인 노드($n_1$)가 0개 또는 1개만 존재할 수 있기 때문에 이를 종합하여 계산하면, 전체 노드 수가 $n$일 때 리프 노드의 수는 항상 $\lceil n / 2 \rceil$개가 된다.

2) 위 1)에서 리프 노드의 수와 내부 노드의 수는 어떤 관계인가? 답: 리프 노드의 수는 내부 노드의 수보다 항상 1개 더 많거나 같습니다.          
($n_L = n_I + 1$ 또는 $n_L = n_I$)  / 일반적인 이진 트리의 성질은 자식이 2개인 노드($n_2$)와 리프 노드($n_0$) 사이에 항상 $n_0 = n_2 + 1$이라는 관계가 있다. 완전 이진 트리에서의 적용은 내부 노드($n_I$)는 $n_2 + n_1$으로 구성된다. 만약 자식이 1개인 노드($n_1$)가 0개라면  $n_L = n_I + 1$이 성립된다. 만약 자식이 1개인 노드($n_1$)가 1개라면 $n_L = n_I$가 성립한다.결론적으로, 완전 이진 트리에서 리프 노드의 수는 내부 노드의 수와 같거나 1개 더 많은 관계를 가진다.

4) 이진 검색 트리는 다음 성질을 만족한다.
임의의 노드 x에 대하여 왼쪽 서브트리에 있는 모든 키는 x의 키보다 작고
오른쪽 서브트리에 있는 모든 키는 x의 키보다 크다. 중위 순회는 왼쪽 서브트리 방문-> 현재 노드 방문 ->오른쪽 서브트리 순서로 노드를 방문한다.

   이제 트리의 크기 n에 대해 수학적 귀납법으로 증명한다.

    (1)
    노드가 1개인 트리는 중위 순회를 하면 그 노드 하나만 방문하므로, 결과는 자명하게 정렬되어 있다.

    (2)
    노드의 개수가 n−1 이하인 모든 이진 검색 트리에 대해, 중위 순회 결과가 정렬된 순서라고 가정한다.

    (3)
    노드의 개수가 n인 이진 검색 트리를 생각하자.
    루트 노드를 r, 왼쪽 서브트리를 L, 오른쪽 서브트리를 R이라 하자.

    BST의 성질에 의해        
    - L의 모든 키 < r의 키       
    - R의 모든 키 > r의 키         
    귀납 가정에 의해                   
    - L을 중위 순회하면 정렬된 순서     
    - R을 중위 순회해도 정렬된 순서

    중위 순회 결과는 (L의 결과)→r→(R의 결과)

    이때 L의 모든 값은 r보다 작고 R의 모든 값은 r보다 크므로 전체 결과 역시 작은 값 → 큰 값 순서로 정렬된다.

    (4)
    수학적 귀납법에 의해, 모든 이진 검색 트리에 대해 중위 순회를 수행하면 키가 항상 정렬된 순서로 방문됨이 증명된다.

이진 검색 트리의 높이는 트리의 구조에 따라 달라진다.

(1) 최대 높이
최악의 경우, 트리가 한쪽으로만 치우친 형태가 된다. 이 경우 각 노드가 하나의 자식만 가지므로, 연결 리스트와 같은 구조가 된다.

따라서 높이는


$$
h = n - 1
$$

이 된다. (노드 수가 n일 때, 간선 기준 높이)

(2) 최소 높이
최선의 경우, 트리가 완전히 균형 잡힌 형태(완전 이진 트리 또는 포화에 가까운 구조)가 된다.이 경우 각 레벨마다 최대한 많은 노드를 가지게 된다.

높이는 다음과 같이 표현된다.

최소 높이:

$$
h = \lfloor \log_2 n \rfloor
$$


(또는 정의에 따라 $$h = \lceil \log_2 (n+1) \rceil - 1$$ 가능)




(3) 결론

- 최대 높이: $O(n)$  
- 최소 높이: $O(\log n)$

즉, 이진 검색 트리의 높이는 트리의 균형 상태에 따라 선형 시간부터 로그 시간까지 달라진다.
